```python
# leitura do arquivo de dados do PBE-VEICULAR
# salvando dados em uma lista para facilitar a manipulação

dados = []

caminho = "2021.txt"

with open(caminho, "r") as arquivo:
  for linha in arquivo:
    dados.append(linha)
```

```python
# verifica o tamanho do arquivo de dados

len(dados)
```

```python
# verifica as possibilidades de tamanhos no arquivo

lista_tamanhos = []

for linha in dados:
  tamanho = linha.split(" ")
  condicao = len(tamanho)
  lista_tamanhos.append(condicao)

set(lista_tamanhos)
```

```python
# listas vazias que representa a quantidade de items dentro de cada linha
# após identificar padrões de tamanho será verificado se existe um comportamento
# ou relação com o conjunto de dados


lista_24 = []
lista_25 = []
lista_26 = []
lista_27 = []
lista_28 = []
lista_29 = []
lista_30 = []
lista_31 = []
lista_32 = []
```

```python
# laço for que identifica a quantidade de items nas listas e segrega os dados
# com semelhanças em tamanho de items


for linha in dados:
  dados = linha.split(" ")
  condicao = len(dados)
  if condicao == 24:
    lista_24.append(dados)
  
  elif condicao == 25:
    lista_25.append(dados)

  elif condicao == 26:
    lista_26.append(dados)

  elif condicao == 27:
    lista_27.append(dados)
  
  elif condicao == 28:
    lista_28.append(dados)
  
  elif condicao == 29:
    lista_29.append(dados)
  
  elif condicao == 30:
    lista_30.append(dados)

  elif condicao == 31:
    lista_31.append(dados)

  elif condicao == 32:
    lista_32.append(dados)
```
